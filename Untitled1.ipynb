```python
def gamma_kappa_dis(gamma_kappa):
    try:
        gamma, kappa = gamma_kappa

        sat = 0.0001#0.0000125
        precip = 250
        factor = 0.8
        model = pybamm.lithium_sulfur.ZeroD_Chemistry_5()

        param = model.default_parameter_values
        param.update(
            {
                "Current function [A]": 0.0422,
                "Initial Condition for S8 ion [g]": S8_initial_dis/factor,
                "Initial Condition for S6 ion [g]": S6_initial_dis/factor,
                "Initial Condition for S4 ion [g]": S4_initial_dis/factor,
                "Initial Condition for S2 ion [g]": S2_initial_dis/factor,
                "Initial Condition for S ion [g]": S1_initial_dis/factor,
                "Initial Condition for Precipitated Sulfur [g]": Sp_initial_dis/factor,
                "Initial Condition for Terminal Voltage [V]": V_initial_dis,
                "Standard Potential L [V]": 1.9,#1.9,#2.0,
                "Standard Potential M [V]": 1.95,#1.95,#2.0,
                "Standard Potential H [V]": 2.38,#2.35,#2.38,
                "Lower voltage cut-off [V]" : 1.8,
                "Shuttle rate coefficient during discharge [s-1]": 0,#.000002,
                "Precipitation rate [s-1]": precip,
                "S saturation mass [g]": sat,
                "gamma": gamma,
                "kappa": kappa
            }
        )


        sim = pybamm.Simulation(
            model,
            #experiment=experiment,
            parameter_values = param,
            solver=pybamm.CasadiSolver(
                atol=1e-6,
                rtol=1e-3,
                dt_max = .01,
                extra_options_setup={"max_step_size": 0.1}, 
                max_step_decrease_count = 15,
            ),
        )
        hours = 4.8
        time_in_seconds = 2000#3500#16950#3600*hours
        sim.solve(np.linspace(0, time_in_seconds, int(time_in_seconds)))

        condition = True 
        while condition:
            try:
                sim.step(100)
                if sim.solution.t[-1] > 13000:#16500:
                    condition = False
                if sim.solution["Terminal voltage [V]"].entries[-1] <1.8:
                    condition = False
            except:
                condition = False

        condition = True 
        while condition:
            try:
                sim.step(50)
                if sim.solution.t[-1] > 13000:#16500:
                    condition = False
                if sim.solution["Terminal voltage [V]"].entries[-1] <1.8:
                    condition = False
            except:
                condition = False

        condition = True 
        while condition:
            try:
                sim.step(10)
                if sim.solution.t[-1] > 13000:#16500:
                    condition = False
                if sim.solution["Terminal voltage [V]"].entries[-1] <1.8:
                    condition = False
            except:
                condition = False
        return sim
    except:
        pass
```

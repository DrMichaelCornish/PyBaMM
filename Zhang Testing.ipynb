```python
experiment = pybamm.Experiment(
    [
        (
            "Discharge at 1 A for 3.2 hours or until 2.2 V",
            "Charge at 1 A for 10 hours or until 2.4 V"
        ),
    ],
)

model15 = pybamm.lithium_sulfur.ZhangEtAl2015()

params = model15.default_parameter_values
params.update(
    {
        "Zhang Initial Condition for solid S8 [g]": 670,
        "Zhang Initial Condition for S8 ion [g]": 100,
        "Zhang Initial Condition for S6 ion [g]": 8.2,
        "Zhang Initial Condition for S4 ion [g]": .0056,
        "Zhang Initial Condition for S2 ion [g]": .000008,
        "Zhang Initial Condition for S ion [g]": .000000014,
        "Zhang Initial Condition for Reaction 2 overpotential [V]": 2.4,
        "Precipitation rate [s-1]": 0,
        "Initial Specific Surface Area [m-1]" : 1e-12
        
    }
)

sim15 = pybamm.Simulation(
    model15,
    #experiment=experiment,
    parameter_values = params,
    solver=pybamm.CasadiSolver(
        atol=1e-6,
        rtol=1e-3,
        dt_max = 10,
        extra_options_setup={"max_step_size": 0.1}, 
        max_step_decrease_count = 15,
    ),
)
time_in_seconds = 1000
sim15.solve(np.linspace(0,.0018,100))#np.linspace(0, time_in_seconds, int(time_in_seconds/10)))
```
